### Current Community Supervision Offenders Files 

Offender_root: Basic offender information on current community supervision offenders <br>
Offender_aliases: Aliases for offenders <br>
Offender_offenses_CCS: Current community supervision offenses for offenders <br>
Offender_residence: Most recent address (excluding confidential addresses) for offenders

```python
# Offender_root: Basic offender information on current community supervision offenders
df_offender_root = pd.read_excel("../data/source/FL-OBIS/offender/Offender_Apr_2022.XLSX")
```

```python
df_offender_root.sample(5)
```

```python
# Offender_residence: Most recent address (excluding confidential addresses) for offenders
df_offender_residence = pd.read_excel("../data/source/FL-OBIS/offender/Offender_Apr_2022_residence.XLSX")
```

```python
df_offender_residence.sample(5)
```

```python
# Offender_offenses_CCS: Current community supervision offenses for offenders 
df_offender_offenses_css = pd.read_excel("../data/source/FL-OBIS/offender/Offender_Apr_2022_offenses_css.XLSX")
```

```python
df_offender_offenses_css.sample(5)
```

```python
# Offender_aliases: Aliases for offenders
df_offender_alias = pd.read_excel("../data/source/FL-OBIS/offender/Offender_Apr_2022_alias.XLSX")
```

```python
df_offender_alias.sample(5)
```

```python
df_offender = pd.merge(df_offender_root, df_offender_residence, on='DCNumber', how='outer')
df_offender.sample(2)
```
